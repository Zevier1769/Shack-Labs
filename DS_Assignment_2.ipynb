```python
#droping 'crawl_timestamp','product_url', 'product_category_tree', 'image', 'is_FK_Advantage_product', 'description', 'product_rating','overall_rating','brand', 'product_specifications' feature from both csvs
amzn = amzn.drop(['crawl_timestamp','product_url', 'product_category_tree', 'image', 'is_FK_Advantage_product', 'description', 'product_rating','overall_rating','brand', 'product_specifications'], axis =1)
flp = flp.drop(['crawl_timestamp','product_url', 'product_category_tree', 'image', 'is_FK_Advantage_product', 'description', 'product_rating','overall_rating','brand', 'product_specifications'], axis =1)
```

```python
#renaming columns in Amazon csv
amzn.rename(columns = {'product_name':'Product Name in Amazon', 'retail_price':'Retail Price in Amazon','discounted_price':'Discounted Price in Amazon'}, inplace = True)
```

```python
#renaming columns in flipkart csv
flp.rename(columns = {'product_name':'Product Name in Flipkart', 'retail_price':'Retail Price in Flipkart','discounted_price':'Discounted Price in Flipkart'}, inplace = True)
```

```python
#Program to find output based on unique id
uid = str(input('Enter unique id:'))
rslt_df_flp = flp.loc[flp['uniq_id'] == uid]
rslt_df_amzn = amzn.loc[flp['uniq_id'] == uid]
result = pd.concat([rslt_df_flp, rslt_df_amzn], axis=1, join='inner')
result = result.drop(['uniq_id', 'uniq_id','pid','pid'], axis =1)
result
```

```python
#Program to find output based on product name
name = str(input('Enter Name:'))
name = name.lower()
rslt_df_flp2 = flp.loc[flp['Product Name in Flipkart'].str.lower() == name]
rslt_df_amzn2 = amzn.loc[amzn['Product Name in Amazon'].str.lower() == name]
result2 = pd.concat([rslt_df_flp2, rslt_df_amzn2], axis=1, join='inner')
result2 = result2.drop(['uniq_id', 'uniq_id','pid','pid'], axis =1)
result2
```

```python
#Program to find output based on product id
pid = str(input('Enter Product ID:'))
pid = pid.upper()
rslt_df_flp3 = flp.loc[flp['pid'] == pid]
rslt_df_amzn3 = amzn.loc[amzn['pid'] == pid]
result3 = pd.concat([rslt_df_flp3, rslt_df_amzn3], axis=1, join='inner')
result3 = result3.drop(['uniq_id', 'uniq_id','pid','pid'], axis =1)
result3
```
